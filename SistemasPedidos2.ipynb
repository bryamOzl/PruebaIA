```python
from tkinter import Tk, Canvas, Frame, Label, Entry, Button, W, E, Listbox, END

import psycopg2

root = Tk()
root.title("Sistema Pedidos")

def guardar_cliente(codigo, nombre, direccion, telefono, email):
    conn = psycopg2.connect(dbname="SistemaPedidos", user="postgres",
                            password="cuenca", host="localhost", port="5432")
    cursor = conn.cursor()
    query = '''INSERT INTO clientes(codigo, nombre, direccion, telefono, email) VALUES (%s, %s, %s, %s, %s)'''
    cursor.execute(query, (codigo, nombre, direccion, telefono, email))
    print("Fila Insertada")
    conn.commit()
    conn.close()
    
def guardar_cuenta(nTarjeta, saldoDisponible, fk_cliente):
    conn = psycopg2.connect(dbname="SistemaPedidos", user="postgres",
                            password="cuenca", host="localhost", port="5432")
    cursor = conn.cursor()
    query = '''INSERT INTO cuenta (nTarjeta, saldoDisponible,fk_cliente) VALUES (%s, %s, %s)'''
    cursor.execute(query, (nTarjeta, saldoDisponible, fk_cliente))
    print("Cuenta Insertada")
    conn.commit()
    conn.close()
    
# Interfaz
canvas = Canvas(root, height=400, width=400)
canvas.pack()

frame = Frame()
frame.place(relx=0.1, rely=0.1, relwidth=0.8, relheight=0.8)

label = Label(frame, text="Crear Cliente")
label.grid(row=0, column=1)

#Codigo Imput
label = Label(frame, text="Codigo")
label.grid(row=1, column=0)

entry_codigo = Entry(frame)
entry_codigo.grid(row=1, column=1)
entry_codigo.focus()


# Name Input
label = Label(frame, text="Nombre")
label.grid(row=2, column=0)

entry_nombre = Entry(frame)
entry_nombre.grid(row=2, column=1)
entry_nombre.focus()

# Age
label = Label(frame, text="Direccion")
label.grid(row=3, column=0)

entry_direccion = Entry(frame)
entry_direccion.grid(row=3, column=1)

# Address
label = Label(frame, text="Telefono")
label.grid(row=4, column=0)

entry_telefono = Entry(frame)
entry_telefono.grid(row=4, column=1)

# Email
label = Label(frame, text="Email")
label.grid(row=5, column=0)

entry_email = Entry(frame)
entry_email.grid(row=5, column=1)

# Tarjeta
label = Label(frame, text="Tarjeta")
label.grid(row=7, column=0)

entry_tarjeta = Entry(frame)
entry_tarjeta.grid(row=7, column=1)

# SaldoDisponible
label = Label(frame, text="Saldo")
label.grid(row=8, column=0)

entry_saldo = Entry(frame)
entry_saldo.grid(row=8, column=1)

# Button
button = Button(frame, text="Guardar", command=lambda: guardar_cliente(
    entry_codigo.get(), entry_nombre.get(), entry_direccion.get(), entry_telefono.get(), entry_email.get()))
button.grid(row=6, column=1, sticky=W+E)

    
button = Button(frame, text="Guardar Cuenta", command=lambda: guardar_cuenta(
    entry_tarjeta.get(),entry_saldo.get(),entry_codigo.get()))
button.grid(row=9, column=1, sticky=W+E)


root.mainloop()
```
